```python
# Simulating Kaggle's ADK for Colab
class Tool:
    def __init__(self, name, pipeline_or_node):
        self.name = name
        self.pipeline_or_node = pipeline_or_node

class Agent:
    def __init__(self, name, tools=[]):
        self.name = name
        self.tools = tools

    def run(self, task):
        print(f"ğŸ¤– {self.name} is working on: {task}")
        result = task
        for tool in self.tools:
            result = tool.pipeline_or_node(result)
        return result
```

```python
# Each tool = one task
def research_task(topic):
    return f"Researching '{topic}'... collected 3 insights about recent trends."

def analysis_task(data):
    return f"Analyzing data â†’ AI adoption in healthcare increasing rapidly. ğŸ“Š"

def report_task(analysis):
    return f"ğŸ“„ FINAL REPORT\n{analysis}\nâœ… Conclusion: AI has high potential to improve diagnostics."
```

```python
# Convert functions to Tool objects
research_tool = Tool("ResearchTool", research_task)
analysis_tool = Tool("AnalysisTool", analysis_task)
report_tool   = Tool("ReportTool", report_task)

# Agents for each stage
research_agent = Agent("ResearchAgent", [research_tool])
analysis_agent = Agent("AnalysisAgent", [analysis_tool])
report_agent   = Agent("ReportAgent", [report_tool])
```

```python
def sequential_workflow(topic):
    step1 = research_agent.run(topic)
    step2 = analysis_agent.run(step1)
    step3 = report_agent.run(step2)
    return step3

output = sequential_workflow("AI in Healthcare")
print("\nâœ… SEQUENTIAL OUTPUT:\n", output)
```

```python
from concurrent.futures import ThreadPoolExecutor

# Two agents collect data simultaneously
research_agent_1 = Agent("ResearchAgent1", [research_tool])
research_agent_2 = Agent("ResearchAgent2", [research_tool])

def parallel_workflow(topic):
    with ThreadPoolExecutor() as executor:
        results = list(executor.map(lambda t: research_task(t),
                                    [topic + " (papers)", topic + " (statistics)"]))
    combined_data = " | ".join(results)
    analysis = analysis_agent.run(combined_data)
    report = report_agent.run(analysis)
    return report

output_parallel = parallel_workflow("AI in Healthcare")
print("\nâš¡ PARALLEL OUTPUT:\n", output_parallel)
```

```python
def loop_workflow(topic, max_loops=3):
    for i in range(max_loops):
        print(f"\nğŸ”„ Iteration {i+1}")
        report = sequential_workflow(topic)
        if "Conclusion" in report:
            print("âœ… Report meets quality check!")
            return report
    return report

final_output = loop_workflow("AI in Healthcare")
print("\nğŸ FINAL OUTPUT:\n", final_output)
```

```python
topic = input("Enter a topic to research: ")
result = loop_workflow(topic)
print("\nğŸ“˜ AGENT TEAM RESULT:\n", result)
```
